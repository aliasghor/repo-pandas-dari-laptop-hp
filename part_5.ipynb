```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Persiapan Data Frame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)),columns=cols)
df['teks'] = ['EMP 1','EMP 2','EMP3','EMP4','EMP5']
df
```

# Ngebalik Urutan Kolom

```python
df.loc[:,::-1]
```

# Ngebalik Urutan Baris

```python
df.loc[::-1]
```

# Ngebalik urutan Baris Dan Melakukan Penyesuaian Ulang Index

```python
df.loc[::-1].reset_index(drop=True)
```

```python
df.plot(kind='line',x='teks',y='A',marker='s',linestyle='-',c='r')

plt.title('Line Plot')
plt.ylabel('Jumlah Karyawan')
plt.xlabel('Karyawan')

plt.grid()
plt.show()
```

```python
ax = plt.gca()

df.plot(kind='line',x='teks',y='A',ax=ax,marker='o',c='r',linestyle='--')
df.plot(kind='line',x='teks',y='B',ax=ax,marker='s',c='b',linestyle='-')

plt.title('Multiple Line Plot')
plt.xlabel('Karyawan')
plt.ylabel('Jumlah Karyawan')

plt.grid()
plt.show()
```

```python
df.plot(kind='bar',x='teks',y='C')

plt.title('Multiple Line Plot')
plt.xlabel('Karyawan')
plt.ylabel('Jumlah Karyawan')

plt.show()
```

```python
df.plot(kind='barh',x='teks',y='C')


plt.title('Multiple Line Plot')
plt.xlabel('Karyawan')
plt.ylabel('Jumlah Karyawan')

plt.show()
```

```python
x = np.linspace(1,10,100)

s = np.sin(2 * np.pi * x)

s
```

```python
fig, ax = plt.subplots()

ax.plot(x,s)

ax.set(xlabel='nilai x',ylabel='nilai sin(x)',title='visualisasi nilai sin')

ax.grid()
plt.show()
```
