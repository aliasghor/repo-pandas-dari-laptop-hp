```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols)
df
```

# Selection Dengan Operator Logika |

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

# Selection Dengan Isin()

```python
df[df['A'].isin([1,3])]
```

# Operator Negasi ~

```python
df[~df['A'].isin([1,3])]
```
